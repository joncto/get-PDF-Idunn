# Funksjon for å laste ned URLer

Argumentet nedenfor er en iterabel der det første elementet er URL som skal lastes ned, og det andre elementet er filnavn. Elementene tilegnes variabler  (url og fn) for lesbarhet.

```python
def download_url(args):
    t0 = time.time()
    url, fn = args[0], args[1]
    try:
        r = requests.get(url)
        with open(fn, 'wb') as f:
            f.write(r.content)
        return(url, time.time() - t0)
    except Exception as e:
        print('Exception in download_url():', e)
```

# Parallell nedlasting

Koden nedenfor definerer funksjonen for parallell nedlasting.

`imap_unordered` sendes til `download_url`-funksjonen, og `input` argumenter til `download_url` (input-variablene). Ved å kjøre `download_url` parallelt minimeres prosessorbruk og dermed den totale nedlastningstiden.

```python
def download_parallel(args):
    cpus = cpu_count()
    results = ThreadPool(cpus - 1).imap_unordered(download_url, args)
    for result in results:
        print('url:', result[0], 'time (s):', result[1])
```

```python
download_parallel(inputs)
```
